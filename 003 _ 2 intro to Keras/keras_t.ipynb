```python
from keras.models import Sequential 
from keras.layers import Input,Dense
import numpy as np 

model = Sequential() 
model.add(Input(shape=(1,)))

model.add(Dense(units=1,activation='linear'))

model.compile(optimizer='sgd',loss='mean_absolute_error',metrics=['mae','mape'])


x_train = np.array([1, 2, 3, 4, 5]) # Convert to NumPy array
y_train = np.array([2, 4, 6, 8, 10]) # Convert to NumPy array

epochs = 1000 
batch_size = 2 

model.fit(x=x_train,y=y_train,epochs=epochs,batch_size=batch_size)

```

```python
Weight,bias = model.get_weights()
print(f"Update W =  {round(Weight[0][0],1)}   b ={bias[0]}")
```

```python
x_test = np.array([6,8,10])
pred = model.predict(x_test)
```

```python
import matplotlib.pyplot as plt

model = Sequential() 

# Specify input_shape directly in the first Dense layer
model.add(Dense(units=1, activation='linear', input_shape=(1,)))



model.compile(optimizer='sgd',loss='mean_absolute_error',metrics=['mae','mape'])


x_train = np.array([1, 2, 3, 4, 5]) # Convert to NumPy array
y_train = np.array([2, 4, 6, 8, 10]) # Convert to NumPy array

epochs = 1000 
batch_size = 2 

history = model.fit(x=x_train,y=y_train,epochs=epochs,batch_size=batch_size,verbose=0)



```

```python
plt.figure(figsize=(10,7))
plt.plot(range(epochs),history.history['loss'])
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training Loss')

plt.show()
```

```python
plt.figure(figsize=(10,7))
plt.plot(range(epochs),history.history['mae'])
plt.xlabel('Epochs')
plt.ylabel('Loss (MAE)')
plt.title('Training Loss (MAE)')

plt.show()
```
